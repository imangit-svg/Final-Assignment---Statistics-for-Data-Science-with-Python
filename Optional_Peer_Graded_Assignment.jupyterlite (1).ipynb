```python
import pandas as pd
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as pyplot
import scipy.stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
from js import fetch
import io

URL = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-ST0151EN-SkillsNetwork/labs/boston_housing.csv'
resp = await fetch(URL)
boston_url = io.BytesIO((await resp.arrayBuffer()).to_py())
```

```python
boston_df=pd.read_csv(boston_url)
```

```python
boston_df.describe()
```

```python
boston_df.head(10)
```

```python
ax = sns.boxplot(y = 'MEDV', data = boston_df)
ax.set_title('Owner-occupied homes')
```

```python
boston_df.info()
boston_df.describe()
```

```python
ax1 = sns.countplot(x='CHAS', data=boston_df)  
ax1.set_title('Number of Homes Near the Charles River')
ax1.set_xlabel('Bounded by Charles River (0 = No, 1 = Yes)')
ax1.set_ylabel('Number of Homes')
```

```python
ax1 = sns.scatterplot(x='AGE', y='MEDV', data=boston_df)  # Scatter plot for AGE vs MEDV
ax1.set_title('Median Value of Houses vs Proportion Built Before 1940')
ax1.set_xlabel('Proportion of Houses Built Before 1940 (%)')
ax1.set_ylabel('Median Value (in $1000s)')
```

```python
ax2 = sns.scatterplot(x='NOX', y='INDUS', data=boston_df)  # Scatter plot for NOX vs INDUS
ax2.set_title('Relationship Between Nitric Oxide and Non-Retail Business Acres')
ax2.set_xlabel('Nitric Oxide Concentration (ppm)')
ax2.set_ylabel('Proportion of Non-Retail Business Acres')
```

```python
ax3 = sns.scatterplot(x='DIS', y='MEDV', data=boston_df)  # Scatter plot for DIS vs MEDV
ax3.set_title('Impact of Distance to Employment Centers on Median Home Value')
ax3.set_xlabel('Weighted Distance to Employment Centers')
ax3.set_ylabel('Median Value (in $1000s)')
```

```python
from scipy.stats import ttest_ind
```

```python
river = boston_df[boston_df['CHAS'] == 1]['MEDV']
no_river = boston_df[boston_df['CHAS'] == 0]['MEDV']
t_stat, p_value = ttest_ind(river, no_river)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```

```python
from scipy.stats import pearsonr
```

```python
corr_age_medv, p_value_age_medv = pearsonr(boston_df['AGE'], boston_df['MEDV'])
print(f"Correlation between AGE and MEDV: {corr_age_medv}, P-value: {p_value_age_medv}")
```

```python
corr_nox_indus, p_value_nox_indus = pearsonr(boston_df['NOX'], boston_df['INDUS'])
print(f"Correlation between NOX and INDUS: {corr_nox_indus}, P-value: {p_value_nox_indus}")
```

```python
X = boston_df['DIS']
y = boston_df['MEDV']
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
print(model.summary())
```

```python
boston_df.loc[boston_df['AGE'] <= 35, 'Age_Group'] = '35 years and younger'
boston_df.loc[(boston_df['AGE'] > 35) & (boston_df['AGE'] < 70), 'Age_Group'] = 'between 35 and 70 years'
boston_df.loc[boston_df['AGE'] >= 70, 'Age_Group'] = '70 years and older'
```

```python

ax6 = sns.boxplot(x='Age_Group', y='MEDV', data=boston_df)
ax6.set_title('Median Value of Owner-Occupied Homes by Age Group')
ax6.set_xlabel('Age Group')
ax6.set_ylabel('Median Value (in $1000s)')
```

```python
ax7 = sns.scatterplot(x='NOX', y='INDUS', data=boston_df)
ax7.set_title('Scatter Plot of NOX vs INDUS')
ax7.set_xlabel('Nitric Oxide Concentration (ppm)')
ax7.set_ylabel('Proportion of Non-Retail Business Acres')
```

```python
ax8 = sns.histplot(boston_df['PTRATIO'], bins=10, kde=True)
ax8.set_title('Histogram of Pupil to Teacher Ratio')
ax8.set_xlabel('Pupil to Teacher Ratio')
ax8.set_ylabel('Frequency')
```
